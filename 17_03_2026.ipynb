```python
max_num = None
print("Enter numbers (type 'done' to stop):")

while True:
    entry = input("Enter number: ")

    if entry.lower() == 'done':
        break # Exit the loop if 'done' is entered

    try:
        num = float(entry)
        if max_num is None or num > max_num:
            max_num = num
    except ValueError:
        print("Invalid input. Please enter a valid number or 'done'.")

if max_num is not None:
    print(f"The maximum number entered is: {max_num}")
else:
    print("No valid numbers were entered.")
```

```python
L = list(map(int, input().split()))
n = len(L)
i = 0

if n == 0:
  print("the list is empty")
else:
  max_element = L[0] # Assume the first element is the maximum initially
  for i in range(1, n):
    if L[i] > max_element:
      max_element = L[i]
  print(f"The maximum element in the list is: {max_element}")
  print(f"The list entered is: {L}")
```

```python
class Student:
    def __init__(self, name, rollno):
        self.name = name
        self.rollno = rollno

    def display_info(self):
        print(f"Student Name: {self.name}")
        print(f"Roll Number: {self.rollno}")

student1 = Student("Akshay", 101)
student1.display_info()

student2 = Student("Boss", 102)
student2.display_info()
```

```python
class Employee:
  def __init__(self, emp_id ,name, salary):
    self.name = name
    self.salary = salary
    self.emp_id = emp_id

  def display_info(self):
    print(f"Employee ID: {self.emp_id}, Name: {self.name}, Salary: {self.salary}")

# Data for employees
employee_data = {
    101: {"name": "Akshay", "salary": 50000},
    102: {"name": "Boss", "salary": 60000},
    103: {"name": "Bass", "salary": 70000}
}

# Create Employee objects and store them in a list
employees = []
for emp_id, emp_info in employee_data.items():
  employee_obj = Employee(emp_id, emp_info["name"], emp_info["salary"])
  employees.append(employee_obj)

# Display information for each employee
for emp in employees:
  emp.display_info()
```
